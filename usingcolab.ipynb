```python
def linearSearch(arr,n,key):
  for i in range(n):
    if arr[i]==key:
      return i
  return -1
n = int(input())
arr = list(map(int,input().split()))[:n]
key = int(input())
index =linearSearch(arr,n,key)
if(index == -1):
  print("Not Found")
else:
  print(f'{key} is present at index:{index}')
```
